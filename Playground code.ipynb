```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
data.describe()
```

```python
#converting the defects column into a bolean column
data['defects'] = data['defects'].astype(int)
data.head()
```

```python
#making X_train and y_train matrices for neural network
y_train = data.defects
y_train = np.array(y_train)
y_train = y_train.reshape(-1,1)

#removing defects column from dataset
column_to_remove = 'defects'
X_train = data.drop(columns=column_to_remove)

#Standerdizing the data between 0 and 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(X_train)
df_standardized = pd.DataFrame(scaled_data, columns=X_train.columns)
df_standardized.head()
```

```python
X_train = df_standardized.values
```

```python
#training a neural network
model = tf.keras.Sequential([
    layers.Dense(512, activation='relu', input_shape=(22, )),
    layers.Dense(256, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
#importing test data set
test_data = pd.read_csv('test.csv')
ID = test_data.id
test_data.head()
```

```python
#standardizing the dataset
X_test = scaler.fit_transform(test_data)
df_standardized = pd.DataFrame(X_test, columns=test_data.columns)
df_standardized.head()
```

```python
X_test = df_standardized.values
```

```python
#predicting the values of X_test
predictions = model.predict(X_test)
predictions
```

```python
#converting predicted values into true and false 
threshold = 0.30
flattened_predictions = [item for sublist in predictions for item in sublist]
binary_predictions = [1 if value > threshold else 0 for value in flattened_predictions]
# Create the DataFrame
prediction_data = {'id': ID, 'defects': binary_predictions}
df = pd.DataFrame(prediction_data)
df.head()
```

```python
#saving the predictions in a csv file
file_path = 'My_predictions.csv'
df.to_csv(file_path, index=False)
print(f'Dataframe saved as csv to: {file_path}')
```
